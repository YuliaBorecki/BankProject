```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
data = pd.read_csv('Financial market statistics_Yulia.csv', skiprows=9)
data = data.iloc[1:27,:].transpose()
data.columns = data.iloc[0]
data = data.iloc[1:,:]
data
```

```python
data.info()
```

```python
#replace .. with zero and convert objects to floats
data = data.replace(to_replace='..', value=0)
data = data.astype(float)
data.tail(4)
```

```python
#choosing columns for analysis
b_data = data.iloc[:,[0,1,2,3,9,10,11,12,13,14,23]]
b_data
```

```python
b_data.describe()
```

```python
b_data.corr()
```

```python
# plot the heatmap
corr = b_data.corr()
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```
